```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load online dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print(df.head())
```

(a) Line Plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.plot(df.index[:20], df["sepal_length"][:20], marker='o')

plt.title("Sepal Length Line Plot")
plt.xlabel("Sample Number")
plt.ylabel("Sepal Length (cm)")

plt.xticks(range(0,20,2))
plt.yticks(range(4,9))

plt.annotate("Maximum",
             xy=(14, df["sepal_length"][:20].max()),
             xytext=(10,8),
             arrowprops=dict(facecolor='black'))

plt.grid(True)

plt.savefig("line_plot.png")

plt.show()
```

(b) Bar Plots
(i) Grouped Bar Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    "ML":[80,90,70],
    "DL":[75,85,65],
    "DS":[88,92,72]
}, index=["Rahil","Amir","Ankit"])

data.plot(kind="bar", figsize=(8,5))

plt.title("Grouped Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

(ii) Stacked Bar Plot

```python
data.plot(kind="bar",
          stacked=True,
          figsize=(8,5))

plt.title("Stacked Bar Plot")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

(c) Histogram and Density Plot

```python
plt.figure(figsize=(7,5))

plt.hist(df["sepal_length"],
         bins=10)

plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")

plt.show()
```

Density Plot

```python
plt.figure(figsize=(7,5))

sns.kdeplot(df["sepal_length"],
            fill=True)

plt.title("Density Plot")

plt.show()
```

(d) Scatter Plot

```python
plt.figure(figsize=(7,5))

plt.scatter(df["sepal_length"],
            df["petal_length"])

plt.title("Scatter Plot")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

```python
(e) Box Plot
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="species",
            y="sepal_length",
            data=df)

plt.title("Box Plot of Sepal Length")

plt.show()
```

Result

Successfully created and visualized:

A Line Plot with title, axis labels, ticks, annotations, and saved it to a file.
Grouped and Stacked Bar Plots using a DataFrame.
A Histogram and Density Plot for frequency and distribution analysis.
A Scatter Plot to examine the relationship between two variables.
A Box Plot to visualize data across multiple categorical groups using Matplotlib and Seaborn.
