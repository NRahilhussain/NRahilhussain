(a) Hierarchical Indexing
Aim

To create a Series with hierarchical indexing and access data using partial indexing.

```python
import pandas as pd

# Create MultiIndex Series
data = pd.Series(
    [85, 90, 78, 88, 95, 80],
    index=[
        ["CSE", "AIML", "AIDS", "ECE", "IT", "MBA"],
        ["Rahil", "Aisha", "Anzar", "Naveen", "Kalim", "Aman"]
    ]
)

print("Original Series")
print(data)

# Access outer level
print("\nStudents in CSE")
print(data["CSE"])

# Access inner level
print("\nMarks of Rahil")
print(data["CSE"]["Rahil"])
```

(b) Rearranging Data using stack() and unstack()

```python
import pandas as pd

# Create DataFrame
data = {
    "DS": [85, 90],
    "ML": [80, 95]
}

df = pd.DataFrame(data, index=["Rahil", "Aman"])

print("Original DataFrame")
print(df)

# Stack
stacked = df.stack()

print("\nStacked Data")
print(stacked)

# Unstack
unstacked = stacked.unstack()

print("\nUnstacked Data")
print(unstacked)
```

(c) Merge DataFrames and Combine Overlapping Data

```python
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    "Name": ["Rahil", "Anzar", "Naveen"],
    "Marks": [80, 95, 78]
}, index=[101, 102, 103])

# Second DataFrame
df2 = pd.DataFrame({
    "Branch": ["AIML", "CSE", "MBA"],
    "Marks": [89, 92, 86]
}, index=[101, 102, 103])

print("DataFrame 1")
print(df1)

print("\nDataFrame 2")
print(df2)

# Merge using index
merged = pd.merge(
    df1,
    df2,
    left_index=True,
    right_index=True,
    how="outer",
    suffixes=("_df1", "_df2")
)

print("\nMerged DataFrame")
print(merged)

# Combine overlapping values
combined = df1.combine_first(df2)

print("\nCombined DataFrame using combine_first()")
print(combined)
```

Result
Successfully created a hierarchical index using a MultiIndex Series and accessed data using partial indexing.
Successfully rearranged hierarchical data using the stack() and unstack() methods.
Successfully merged two DataFrames using their indexes as merge keys and combined overlapping data using the combine_first() method.
