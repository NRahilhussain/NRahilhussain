```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2
```

```python
pop_size = 20
generations = 30
mutation_rate = 0.1
```

```python
population = np.random.uniform(-10,10,pop_size)
best_history = []

for g in range(generations):

    fitness = f(population)

    # Selection
    idx = np.argsort(fitness)
    population = population[idx]

    # Crossover
    children = []
    for i in range(pop_size//2):
        p1 = population[i]
        p2 = population[i+1]
        child = (p1+p2)/2
        children.append(child)

    children = np.array(children)

    # Mutation
    mutation = np.random.rand(len(children)) < mutation_rate
    children[mutation] += np.random.normal(0,1,sum(mutation))

    population = np.concatenate((population[:10],children))

    best_history.append(np.min(f(population)))

print("Best solution:",population[np.argmin(f(population))])
```

```python
plt.plot(best_history)
plt.title("GA Convergence")
plt.show()
```
