(a) Data Cleaning using Pandas
Aim

To identify missing values, remove missing values, fill missing values, and remove duplicate records.

```python
import pandas as pd
import numpy as np

# Create DataFrame with missing values and duplicates
data = {
    "Name": ["Rahil", "Amir", np.nan, "naveen", "anzar"],
    "Age": [30, np.nan, 25, 35, 30],
    "Marks": [80, 90, np.nan, 79, 85]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# Identify missing values
print("\nMissing Values")
print(df.isna())

# Drop rows with missing values
print("\nAfter dropna()")
print(df.dropna())

# Fill missing values
filled_df = df.fillna({
    "Name": "Unknown",
    "Age": df["Age"].mean(),
    "Marks": df["Marks"].mean()
})

print("\nAfter fillna()")
print(filled_df)

# Find duplicates
print("\nDuplicate Rows")
print(filled_df.duplicated())

# Remove duplicates
print("\nAfter drop_duplicates()")
print(filled_df.drop_duplicates())
```

(b) Data Transformation using map(), replace(), and rename()

```python
import pandas as pd

data = {
    "Name": ["Rahil", "Anni", "naveen"],
    "Gender": ["M", "F", "M"],
    "Marks": [81, 70, 78]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# map()
df["Gender"] = df["Gender"].map({"M": "Male", "F": "Female"})

print("\nAfter map()")
print(df)

# replace()
df["Marks"] = df["Marks"].replace({70: 80})

print("\nAfter replace()")
print(df)

# rename()
new_df = df.rename(columns={"Marks": "Score"})

print("\nAfter rename()")
print(new_df)

print("\nOriginal DataFrame")
print(df)
```

(c) Detect Outliers using Random Sampling

```python
import pandas as pd
import numpy as np

# Generate normally distributed data
np.random.seed(10)

data = np.random.normal(loc=50, scale=10, size=100)

# Add outliers
data = np.append(data, [100, 110, 120])

df = pd.DataFrame({"Values": data})

print(df.head())

# Calculate Quartiles
Q1 = df["Values"].quantile(0.25)
Q3 = df["Values"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df["Values"] < lower) | (df["Values"] > upper)]

print("\nPossible Outliers")
print(outliers)
```

(d) Text Manipulation using Regular Expressions

```python
import re

text = "mohan\tbabu   uni\nTirupati   19221"

print("Original String")
print(text)

# Split using variable whitespace
words = re.split(r"\s+", text)

print("\nSplit Output")
print(words)

# Find all words
matches = re.findall(r"\w+", text)

print("\nMatching Patterns")
print(matches)
```
