Reading, Writing and Parsing JSON using read_json()

```python
[
    {
        "RollNo":101,
        "Name":"Rahul",
        "Age":20,
        "Marks":85
    },
    {
        "RollNo":102,
        "Name":"Aisha",
        "Age":21,
        "Marks":90
    },
    {
        "RollNo":103,
        "Name":"Rohan",
        "Age":19,
        "Marks":78
    }
]
```

```python
import pandas as pd

data = {
    "RollNo": [101, 102, 103],
    "Name": ["Rahul", "Aisha", "Rohan"],
    "Age": [20, 21, 19],
    "Marks": [85, 90, 78]
}

df = pd.DataFrame(data)

# Save to JSON
df.to_json("students.json", orient="records", indent=4)

# Read it back
df2 = pd.read_json("students.json")

print(df2)
```

```python
import pandas as pd

# Read JSON file
df = pd.read_json("students.json")

print("JSON Data:")
print(df)

# Write DataFrame to JSON
df.to_json("students_output.json", orient="records", indent=4)

print("\nJSON file written successfully.")
```
